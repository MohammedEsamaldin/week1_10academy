```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/drive/MyDrive/cleaned_data.csv',na_values= ['?',None])
df.head(10)
```

```python
df['Average TCP retransmission']=df['TCP UL Retrans. Vol (Bytes)']+df['TCP DL Retrans. Vol (Bytes)']
df['Average RTT']=df['Avg RTT DL (ms)']+df['Avg RTT UL (ms)']
df['Average throughput']=df['Avg Bearer TP DL (kbps)']+df['Avg Bearer TP UL (kbps)']
```

```python
Ex_agg = df.groupby(['MSISDN/Number']).agg({'Average TCP retransmission':'sum','Average RTT':'sum','Average throughput':'sum'})
Ex_agg.head()
```

```python
features = ['Average TCP retransmission','Average RTT','Average throughput']
features_array = Ex_agg.loc[:,features].values
features_array = StandardScaler().fit_transform(features_array)
features_array
```

```python
normalized_data = pd.DataFrame(features_array,columns=features)
normalized_data
```

```python
kmeans = KMeans(3)
kmeans.fit(features_array)
clusters = kmeans.fit_predict(features_array)
centroids = kmeans.cluster_centers_
clusters
```

```python
normalized_data['cluster_num'] = clusters
normalized_data
```

```python
non_nomalized_data = Ex_agg[features]
non_nomalized_data['cluster'] = clusters
non_nomalized_data.head(10)
```

```python
satisfaction_data = non_nomalized_data.groupby('cluster').agg({'Average TCP retransmission':'mean','Average RTT':'mean','Average throughput':'mean'})
satisfaction_data.head()
```
